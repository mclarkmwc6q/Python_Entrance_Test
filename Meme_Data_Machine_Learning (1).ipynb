```python
from sklearn import tree
X = [[1,0.02,0.3],
[1,0.01,0.15],
[0,0.03,0.15],
[0,0.03,0.3],
[0,0.01,0.15],
[0,0.02,0.15],
[0,0.01,0.45],
[0,0.01,0.6],
[1,0.03,0.3],
[1,0.02,0.45],
[1,0.01,0.6],
[0,0.04,0.3],
[1,0.04,0.15],
[0,0.02,0.45],
[0,0.02,0.3]]
Y = [0,0,0,0,1,1,0,0,1,1,1,0,1,0,1]
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, Y)
```

```python
clf.predict([[0,0.01,0.3],
[1,0.03,0.45],
[1,0.02,0.15],
[0,0.04,0.45],
[1,0.03,0.15]])
```
